# Objective

The purpose of this project is to analyze and clean adequate data about dermatoscopic lesions. There are three steps to achieve that:

1) Import all the data (images and metadata);

2) Prepare the data. I will be using two methods for that:

    -classic method, where the features and biases are removed through scientific weight (E.g the localization of the lesion does not affect the visual strcuture of the abrasion) and/or visual analysis;
    
    -PCA (Principal Component Analysis) to decrease the dimensions of the dataframe from the default amount to 2 dimensions. Hopefully this will result with a more accurate machine learning model and helps to fit the model more quickly.
    
3) Use machine learning to fit the model and finally calculate it's score, which reflects the accuray of our model. The idea and hope is to prepare the data and fit the model in fashion, which would result us with a score of 95+ percentile.

# Data import

## Import

Import the metadata and sort the "image_id" colum in ascending order

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
metadata =  pd.read_csv("HAM10000_metadata.csv")
metadata = metadata.sort_values("image_id", ascending=True)
metadata.head()
```

Import the image data, make it grayscale and find the matrix value for each image. 

```python
#from IPython.display import Image
from PIL import ImageOps, Image
img_gray = []
img_matrix = []
counter = 0
for i in range(24306, 34321):
    img = Image.open('ISIC_00'+str(i)+'.jpg')
    img = ImageOps.grayscale(img)
    img_gray.append(img)
    img_matrix.append((np.asarray(img)).reshape(-1))
    counter += 1
    
```

Create a dataframe for the newfound array list.

```python
matrices = pd.DataFrame(data={"img_matrix":img_matrix})
matrices.head()
```

First, I'll reset the metadata index values in order to replace the correct values with correct images. Then I'll replace the "image_id" column values with the "img_matrix" column values.

```python
metadata = metadata.reset_index()
metadata["image_id"] = matrices["img_matrix"]
metadata.head()
```

Let's use the Principal Component Analysis (PCA) for the array, in order to find relevant and usable values for fitting the Random Forest Classifier. 

I'll be using the StandardScaler library to find the PCA values and the following steps: 

    a)First, we need to meet the StandardScaler requirements by standardizing the data 
      (applying the method "reshape" to the array is perfect for that);
    b) Second, we need to fit the data and transform it by using the method "fit_transform" 
       for the StandardScaler library;
    c) Third, we need to determine the amount of components required (one in our case);
    d) Fourth, we need to fit and transform the new scaled data (with one component).

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
pca_one = []
pca_two = []
for i in range(len(metadata)):
    shaped = metadata["image_id"].iloc[i].reshape(-1,1)
    scaled = StandardScaler().fit_transform(shaped)

    pca = PCA(n_components=1)
    pca = pca.fit_transform(scaled)
    one = pca[1][0]
    two = pca[0][0]
    pca_one.append(one)
    pca_two.append(two)
    
pca_values = pd.DataFrame(data={"one":pca_one, "two":pca_two})
```

I'll replace the "lesion_id" column with pca_values' column "one" and "image_id" columns with pca_values' column "two".

```python
metadata["lesion_id"] = pca_values["one"]
metadata["image_id"] = pca_values["two"]
```

First, let's randomize the whole dataframe and remove any possible NaN values. There seems to be a bias in some of the features of the dataframe. Let's remedy that by removing some unique values.

**DIAGNOSIS TYPE**

Since all of the diagnoses are 100% accurate and the images are visually consistent, the diagnosis type (column "dx_type") is trivial, and can be completely removed.

**LOCALIZATION**

I will remove the "scalp" value, as the images include lots of hair, which can lead to inaccurate results during analysis. Removal of the bias of different areas of the body is unnecessary, as the structure of the lesions in different locations (both malign and benign) are basically the same.

**SEX**

I'll go ahead and remove the "unknown" value from the column. There are multiple reasons for that: the data is skewed towards values "male" and "female"; a decrease of irrelevant features from the dataframe can lead to more accurate results when fitting the model. 

Caucasian males over the age of 50 have a higher risk of developing melanoma (Source: National Cancer Institute: SEER Stat Fact Sheets: Melanoma of the Skin. Bethesda, Md: National Cancer Institute.), which is why I won't remove the imbalance in the data (where the male count is substancially higher than the female count), as it reflects the natural bias towards males.

**DIAGNOSIS**

Statistically speaking, there is an 20% chance (may vary) for the diagnosed lesion to be a nevus (Source: Population-Based Analysis of Histologically Confirmed Melanocytic Proliferations Using Natural Language Processing; Jason P. Lott, MD, MHS, MSHP1; Denise M. Boudreau, PhD; Ray L. Barnhill, MD, PhD). Since the data is unnaturally skewed towards nevi (value "nv"), I'll decrease the amount of them by ~10%. The reason for it is, that the other diagnoses count doesn't reflect statistical accuracy The removed values from the main dataframe will be added to the "temp_test" dataframe later on.

**AGE**

There seems to be a slight bias towards the middle-age people (45-55). Over half of the melanoma cases and other skin-related issues take place around in the age of 50 (Source: National Cancer Institute: SEER Stat Fact Sheets: Melanoma of the Skin. Bethesda, Md: National Cancer Institute), meaning there is a natural bias towards middle-aged people. Therefore I'll let the bias remain, and not change anything.

# Applying machine learning techniques

```python
#Get dummies for train datasets
#train_img
categories_train_img = []
for i in train_img.columns: #find categorical columns
    if train_img[i].dtypes == "O":
        categories_train_img.append(i) #create a list of all the categorical columns
train_img_dummies = pd.get_dummies(train_img, columns=categories_train_img)

#train_lbl
categories_train_lbl = []
for i in train_lbl.columns: #find categorical columns
    if train_lbl[i].dtypes == "O":
        categories_train_lbl.append(i) #create a list of all the categorical columns
train_lbl_dummies = pd.get_dummies(train_lbl, columns=categories_train_lbl)

#Get dummies for train datasets
#test_img
categories_test_img = []
for i in test_img.columns: #find categorical columns
    if test_img[i].dtypes == "O":
        categories_test_img.append(i) #create a list of all the categorical columns
test_img_dummies = pd.get_dummies(test_img, columns=categories_test_img)

#train_lbl
categories_test_lbl = []
for i in test_lbl.columns: #find categorical columns
    if test_lbl[i].dtypes == "O":
        categories_test_lbl.append(i) #create a list of all the categorical columns
test_lbl_dummies = pd.get_dummies(test_lbl, columns=categories_test_lbl)
```

Although there are many classification algorithms out there, I went for the Random Forest Classifier. The reason for that is, they don't expect linear features or features that interact linearly. They can also handle binary features very well and are great for bagging, which in my case is Decision Tree Classifier*1000 (since the n_estimators=1000 as an RFC parameter). 

Support Vector Machines are great if the problem is not linearly separable, although they are inefficient to train, which is not what I'm looking for.

The best, proven way to achieve great accuracy (in the case of our project) is to use the Convolutional Neural Network (CNN), but since we didn't have time for that RFC will have to do.

```python
from sklearn.ensemble import RandomForestClassifier
sample_data = ['image_id', 'lesion_id', 'age', 'sex', 'localization_abdomen',
       'localization_back', 'localization_chest', 'localization_ear',
       'localization_face', 'localization_foot', 'localization_genital',
       'localization_hand', 'localization_lower extremity',
       'localization_neck', 'localization_trunk', 'localization_unknown',
       'localization_upper extremity']
sample_target = ['dx_akiec', 'dx_bcc', 'dx_bkl', 'dx_df', 'dx_mel', 'dx_nv', 'dx_vasc']
train = RandomForestClassifier(n_estimators=1000).fit(train_img_dummies[sample_data], train_lbl_dummies[sample_target])
score = train.score(test_img_dummies[sample_data], test_lbl_dummies[sample_target])
```

Here's the score of our algorithm, which reflects the AUC curve and also the accuracy of our model.

```python
print("Final score: " + str(score))
```

Let's try to visually interpret the current state of the dataframe. A great way for that is to use Principal Component Analysis. The idea behind it is to reduce the amount of x features to two features, in order to visualize a 2D projection. The visualization should give us an idea how different (or similar) the diagnoses are from eachother, therefore everything should be taken into account (shape, entropy, depth of color, size etc.). If the diagnose types are very similair, they should be clustered together, and on the contrary they should form isolated clusters.

```python
#Same PCA method used as mentioned above
target = train_img_dummies[sample_data].values
label = train_lbl_dummies[sample_target].values
pc = PCA(n_components=2)
PrincipalComponents = pc.fit_transform(target)
pca_df = pd.DataFrame(data = PrincipalComponents, columns=["component 1", "component 2"])
reduced_data = pd.concat([pca_df, train_lbl], axis=1)

#Plotting the data
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Component 1', fontsize = 15)
ax.set_ylabel('Component 2', fontsize = 15)
ax.set_title('Principal Component Analysis', fontsize = 20)

targets = train_lbl["dx"].unique()
colors = "bgrcmykw"
for target, color in zip(targets,colors):
    indicesToKeep = reduced_data['dx'] == target
    ax.scatter(reduced_data.loc[indicesToKeep, 'component 1'], reduced_data.loc[indicesToKeep, 'component 2'], 
               c = color, s = 50, alpha=0.4)
ax.legend(targets)
ax.grid()
plt.show()
```

The result is a mess of points all laying ontop of eachother. Just because there are no distinguishable cluesters in the scatterplot, it doesn't mean the result is negative, as many of the dermatoscopic lesions are quite similar in appearance: round or slightly chaotic shape, slightly lighter around the border of the lesion and opaque in the center. 

To make sure the indistinguishable scatterplot result is not caused by the decrease of the feature from the PCA, let's use the "explained_variance_ratio_" to find the residual accuracy.

```python
print(pc.explained_variance_ratio_[0])
```

The new dataframe ("reduced_data") is ~97.7 as accurate as the initial dataframe ("train_img_dummies" and "train_lbl_dummies"), therefore the given result of the scatterplot is not related to the use of PCA. 

# Conclusion

Although the initial plan was to reach the accuracy of 95+ percentile with the dermatoscopic lesion distinguishing algorithm, we only reached the 62+ percentile. There are multiple reasons for that:
* We used Random Forest Classifier to fit the model, even though similar projects insisted on using Convolutional Neural Network, as it tries to minimize the cost function to categorize the inputs correctly in classification tasks.
* We should've used OpenCV3's Saliency library instead of using grayscale to combat overfitting. Static Saliency gives a more dynamic grayscaled image with more attention to the focus point, which in our case is the lesion. Although good in theory it is quite taxing to the memory of the computer, and is therefore inefficient in our case.
* We should've cropped the images and focus only on the center area (the majority of the images had the lesion in the very center), which would have decreased the noise and the amount of features, resulting with a more efficient model.
* An unseen bias towards nevi. Although a nevus is quite common in a dermatoscopic diagnosis, we did not notice that almost 70% of the data are nevi. It made us ignore the natural bias, which is roughly about 20% of the diagnosed lesion cases, and remain the data skewed.
